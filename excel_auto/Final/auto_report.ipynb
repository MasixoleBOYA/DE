```python
work_book = load_workbook(file_path)

# PART 1: renames the "Gantry AP" column for all sheets
def sheet_rename() -> None:
    for i in work_book.sheetnames:
        work_sheet = work_book[i]
        column_index = 1

        for x in work_sheet.iter_rows(min_row=2, max_row=work_sheet.max_row, min_col=column_index, max_col=column_index):
            for cell in x:
                cell.value = i

# PART 2: Append data from other sheets to a specified sheet
def appending_to_onesheet(sheet_name: str) -> None:
    appending_sheet = work_book[sheet_name]

    for depot_name in work_book.sheetnames[1:]:
        current_worksheet = work_book[depot_name]
        for row in current_worksheet.iter_rows(min_row=2, max_row=current_worksheet.max_row, values_only=True):
            appending_sheet.append(row)



sheet_rename()
appending_to_onesheet("Alrode")
# work_book.save('C:/Users/tshol/Desktop/AAAAAAAA_second_third.xlsx')
```

# 1.1 Attempting to cast Customer codes dictionary into dataframe

# 1.2 Attempting to cast Product codes dictionary into dataframe

```python
# Convert dictionary to DataFrame
df_ProductCodes = pd.DataFrame(list(product_codes_dictionary.items()), columns=['Product Code', 'Product Description'])

# Define the values to be replaced and their corresponding replacements
replacements = {
    'Excellium Unleaded 95': 'ULP 95',
    'Excellium Diesel 50': 'D50',
    'Additised ULP93' :'ULP 93'
}

# Replace the values in the 'Product Description' column
df_ProductCodes['Product Description'] = df_ProductCodes['Product Description'].replace(replacements)

# Print the updated DataFrame
df_ProductCodes


# Print the DataFrame
df_ProductCodes.head()
```

# Merging for the NEW customer names column

```python
# Convert 'Customer No.' column in the second dataframe to numeric, coercing errors
df['Customer No.'] = pd.to_numeric(df['Customer No.'], errors='coerce')

# Merge the two dataframes on the common column 'CustomerID' and 'Customer No.'
df = pd.merge(df, df_CustomerCodes, left_on='Customer No.', right_on='Customer ID', how='left')

# Drop the duplicate 'CustomerID' column from the merged dataframe
df.drop(columns=['Customer ID'], inplace=True)

# Rename the 'CustomerName' column to 'Customer Names'
df.rename(columns={'Customer Name': 'Customer Names'}, inplace=True)

# Print the updated dataframe
df.head()

```

# Merging for new product names column

```python
df['Product'] = pd.to_numeric(df['Product'], errors='coerce')


# Merge the two dataframes on the common column 'Product Code' and 'Product'
df = pd.merge(df, df_ProductCodes, left_on='Product', right_on='Product Code', how='left')


# Drop the duplicate 'Product Code' column from the merged dataframe
df.drop(columns=['Product Code'], inplace=True)

# Print the updated dataframe
df
```
